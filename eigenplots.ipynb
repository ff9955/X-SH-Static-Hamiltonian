```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lattice = 5

l = [1, 0.9, 0.8]
l2 = [0.8, 0.9, 1]

p = [0.3, 0.4, 0.1]

fig, axs = plt.subplots(ncols = 2, nrows = 1)

axs[0].imshow(np.array([l]), cmap="Reds")
axs[0].yaxis.set_visible(False)
#axs[0].xaxis.set_visible(False)

for index in range(len(p)):
    axs[0].scatter(index, 0, c = 'b', s = 1500*(p[index]))

axs[1].imshow(np.array([l2]), cmap="Blues")
axs[1].yaxis.set_visible(False)
#axs[1].xaxis.set_visible(False)

plt.subplots_adjust(wspace=0, hspace=0)
```

```python
fig, axs = plt.subplots(nrows = 10, ncols = 2, sharex = True, figsize = (4,5.5))

for j in range(10):

    l = np.random.rand(1,5)
    l2 = np.random.rand(1,5)
    l3 = np.random.rand(5)

    axs[j,0].imshow(l, cmap = 'Blues')
    #axs[j,0].yaxis.set_visible(False)
    axs[j,0].xaxis.set_visible(False)
    axs[j,0].set_yticks([0])
    axs[j,0].set_yticklabels([int(np.sum(l))])

    for i in range(len(l3)):

        axs[j,0].scatter(i, 0, c = 'm', s = int(100*l3[i]))

    axs[j,1].imshow(l2, cmap = 'Reds')
    axs[j,1].yaxis.set_visible(False)
    axs[j,1].xaxis.set_visible(False)

plt.subplots_adjust(wspace = 0, hspace = 0)
plt.ylabel('Energy')

fig.supylabel('Eigenstate Energy /meV')
```
